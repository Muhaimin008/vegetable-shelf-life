```python
import pandas as pd
import matplotlib.pyplot as plt

# Assuming you have loaded your data into a DataFrame called df

# Apply moving average filter
window_size = 3  # Adjust the window size as needed
smoothed_temperature = df['temperature'].rolling(window=window_size, min_periods=1, center=True).mean()

# Plot original temperature data
plt.figure(figsize=(20, 10))
plt.xlabel('Date Time')
plt.ylabel('Temperature')
plt.plot(df['datetime'], df['temperature'], label='Original')

# Plot smoothed temperature data
plt.plot(df['datetime'], smoothed_temperature, label=f'Smoothed (Window Size={window_size})')

# Add legend
plt.legend()

# Show the plot
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Assuming you have loaded your data into a DataFrame called df

# Filter CO2 data based on the threshold
threshold = 5
df_filtered = df[df['co2'] <= threshold]

# Apply moving average filter to the filtered CO2 data
window_size = 5  # Adjust the window size as needed
smoothed_co2 = df_filtered['co2'].rolling(window=window_size, min_periods=1, center=True).mean()

# Plot CO2 data
plt.figure(figsize=(15, 12))
plt.xlabel('Date Time')
plt.ylabel('CO2')

# Plot original CO2 data
plt.plot(df_filtered['datetime'], df_filtered['co2'], label='Original')

# Plot smoothed CO2 data
plt.plot(df_filtered['datetime'], smoothed_co2, label=f'Smoothed (Window Size={window_size})')

# Add legend
plt.legend()

# Show the plot
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from scipy.signal import medfilt

# Assuming you have loaded your data into a DataFrame called df

# Filter CO2 data based on the threshold
threshold = 5
low_val= 0
df_filtered = df[(df['co2'] <= threshold) & (df['co2'] >= low_val)]

# Apply median filter to the filtered CO2 data
window_size = 3  # Adjust the window size as needed
smoothed_co2 = medfilt(df_filtered['co2'], kernel_size=window_size)

# Plot CO2 data
plt.figure(figsize=(20, 10))
plt.xlabel('Date Time')
plt.ylabel('CO2')

# Plot original CO2 data
plt.plot(df_filtered['datetime'], df_filtered['co2'], label='Original')

# Plot smoothed CO2 data
plt.plot(df_filtered['datetime'], smoothed_co2, label=f'Smoothed (Window Size={window_size})')

# Add legend
plt.legend()

# Show the plot
plt.grid(True)
plt.tight_layout()
plt.show()
```
